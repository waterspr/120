### 2. 删除空值行

```python
jz = jianzhu.dropna(how='any') #全部空值就删去
jz.head()
```

```python
# jz.columns = jz.iloc[0]
jz.rename(columns=jz.iloc[0],inplace=True)   #指定列名
jz.head()
```

#### set_index( ) 将 DataFrame 中的列转化为行索引，
此处drop=False，的作用，是否保留原列
DataFrame可以通过set_index方法，可以设置单索引和复合索引。
DataFrame.set_index(keys, drop=True, append=False, inplace=False, verify_integrity=False) 

append添加新索引，drop为False，inplace为True时，索引将会还原为列

```python
jz = jz.set_index('指标')
jz.head()
```

```python
jz = jz.drop(['指标'])  #删去 指定行
jz.head()
```
