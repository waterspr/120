# pyecharts基本图表

## 日历图

```python
import pandas as pd
import numpy as np
```

```python
import datetime
import random

from pyecharts import options as opts
from pyecharts.charts import Calendar
```

```python
#设置时间和数据\
begin = datetime.date(2019,1,1)
end = datetime.date(2019,6,30)
data = [    
     [str(begin + datetime.timedelta(days=i)), random.randint(1000, 25000)]
         for i in range((end - begin).days + 1) ]
```

```python
data
```

```python
c  = (   
    Calendar()   
    .add("什么", data, calendar_opts=opts.CalendarOpts(range_=['2019-01', '2019-07']))
    .add("事情", data, calendar_opts=opts.CalendarOpts(range_=['2019-01', '2019-07']))   #数据，时间范围
   .set_global_opts(   title_opts=opts.TitleOpts(title="Calendar-2019年上半年微信步数情况"),
                    visualmap_opts=opts.VisualMapOpts( max_=25000, min_=500,orient="horizontal",is_piecewise=True, pos_top="230px",         
                                                      pos_left="60px", ),
                   )
)

c.render_notebook()
```

## Funnel：漏斗图

```python
#导入库\
from pyecharts.faker import Faker  #生成假数据的
from pyecharts import options as opts
from pyecharts.charts import Funnel, Page
```

```python
for z in zip(Faker.choose(), Faker.values()):
    print(z)
    print(type(z))
    print(list(z))  #将元组数据转为list数据\n
```

```python
[list(z) for z in zip(Faker.choose(), Faker.values())]
```

```python
data1= [list(z) for z in zip(Faker.choose(), Faker.values())]
c = (   Funnel()
     .add("商品", data1)
     .set_global_opts(title_opts=opts.TitleOpts(title="Funnel-基本示例"))
    )

c.render_notebook()
```

```python
c = (
    Funnel()
    .add("商品", data1 ,label_opts=opts.LabelOpts(position="inside"),)
    .set_global_opts(title_opts=opts.TitleOpts(title="Funnel-基本示例"))
                                              )
c.render_notebook()
```

```python
c = ( Funnel()
     .add("商品", data1 ,sort_="ascending",  label_opts=opts.LabelOpts(position="inside"),)
    .set_global_opts(title_opts=opts.TitleOpts(title="Funnel-基本示例"))
                                              )

c.render_notebook()
```

## Gauge：仪表盘

```python
from pyecharts import options as opts
from pyecharts.charts import Gauge, Page
```

```python
c = ( 
    Gauge()
     .add(
         "业务指标",[("完成率", 55.5)],
         split_number=5,
         axisline_opts=opts.AxisLineOpts(  #坐标圆环颜色
            linestyle_opts=opts.LineStyleOpts(
                color=[(0.3, "#67e0e3"), (0.7, "#37a2da"), (1, "#fd666d")], width=30
                       ),
                 )
     )
    .set_global_opts(
            title_opts=opts.TitleOpts(title="Gauge-分割段数-Label"),
            legend_opts=opts.LegendOpts(is_show=False),  )
   ) 
                       
c.render_notebook()
```
